```python
import matplotlib.pyplot as plt
import numpy as np

# Generate x values
x = np.linspace(-5, 5, 100)  # 100 points from -5 to 5

# Define the equations of the lines
y1 = 2*x + 1
y2 = 2*x + 2
y3 = 2*x + 3

# Create the plot
plt.figure(figsize=(10, 6))

# Plot each line with different styles and colors
plt.plot(x, y1, 'r--', label='y = 2x + 1')  # Red dashed line
plt.plot(x, y2, 'b:', label='y = 2x + 2')   # Blue dotted line
plt.plot(x, y3, 'g-.', label='y = 2x + 3')  # Green dash-dot line

# Set the title and labels
plt.title('Plot of Linear Equations')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show the legend
plt.legend()

# Display the plot
plt.grid()  # Add grid for better visualization
plt.show()
```

```python
import numpy as np

# Step 1: Define the matrix A
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

# Step 2: Calculate the inverse of matrix A
A_inv = np.linalg.inv(A)

# Step 3: Verify that A * A_inv and A_inv * A produce the identity matrix
identity1 = np.dot(A, A_inv)
identity2 = np.dot(A_inv, A)

# Display the results
print("Matrix A:")
print(A)
print("\nInverse of Matrix A:")
print(A_inv)
print("\nProduct of A and A_inv (A * A_inv):")
print(identity1)
print("\nProduct of A_inv and A (A_inv * A):")
print(identity2)

# Step 4: Optionally, round the results to a few decimal places for clearer output
identity1_rounded = np.round(identity1, decimals=6)
identity2_rounded = np.round(identity2, decimals=6)

print("\nRounded Product of A and A_inv (A * A_inv):")
print(identity1_rounded)
print("\nRounded Product of A_inv and A (A_inv * A):")
print(identity2_rounded)
```
